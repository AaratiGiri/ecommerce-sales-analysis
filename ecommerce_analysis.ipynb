```python
import pandas as pd
from sqlalchemy import create_engine
from sqlalchemy.engine import URL

# Step 1 - Load CSV
cleaned_ecommerce = pd.read_csv('cleaned_ecommerce.csv')
print("✅ CSV Loaded! Rows:", len(cleaned_ecommerce))
data = pd.read_csv('data.csv', encoding='latin1')
print("✅ CSV Loaded! Rows:", len(data))



# Step 2 - Connect to PostgreSQL
connection_url = URL.create(
    drivername="postgresql+psycopg2",
    username="postgres",
    password="@mami1515",
    host="localhost",
    port=5432,
    database="ecommerce"
)

engine = create_engine(connection_url)
print("✅ Connected to PostgreSQL!")

# Step 3 - Export to PostgreSQL
cleaned_ecommerce.to_sql(
    'ecommerce',          # table name in PostgreSQL
    engine,
    if_exists='replace',  # replace if table already exists
    index=False           # don't export dataframe index
)
data.to_sql(
    'original_data',
    engine,
    if_exists='replace',
    index = False
)


print("✅ Data exported successfully!")
print("Rows exported:", len(cleaned_ecommerce))
```
